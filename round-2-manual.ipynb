```python

import matplotlib.pyplot as plt
import numpy as np

x = [x for x in range(101)]
Research = [200_000 * np.log(1 + x_value) / np.log(1 + 100) for x_value in x]
Scale = [ x_value * ( 7 / 100) for x_value in x]

derivative = [200_000 / ((x_value + 1) * np.log(101)) for x_value in x]

# for scale in Scale :
#   for x in range(1, 101):
#     pass

print (200_000 * np.log(1 + 1) / np.log(1 + 100))
print (200_000 * np.log(1 + 2) / np.log(1 + 100))
print (200_000 * np.log(1 + 3) / np.log(1 + 100))
print (200_000 * np.log(1 + 4) / np.log(1 + 100))
print (200_000 * np.log(1 + 5) / np.log(1 + 100))
print (200_000 * np.log(1 + 6) / np.log(1 + 100))
print (200_000 * np.log(1 + 7) / np.log(1 + 100))
print (200_000 * np.log(1 + 8) / np.log(1 + 100))
print (200_000 * np.log(1 + 9) / np.log(1 + 100))
print (200_000 * np.log(1 + 10) / np.log(1 + 100))

#Draw linear 
plt.figure(figsize=(15, 7))
plt.plot(x, Research, color='royalblue', linewidth=1, alpha=0.8)
plt.plot(x, Scale, color='green', linewidth=1, alpha=0.8)
plt.show()
```
